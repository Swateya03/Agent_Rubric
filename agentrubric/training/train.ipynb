```python
from google.colab import files
import os

print('Upload train.jsonl and eval.jsonl')
uploaded = files.upload()

os.makedirs('data', exist_ok=True)
for filename in uploaded.keys():
    os.rename(filename, f'data/{filename}')

print(f'Files uploaded: {list(uploaded.keys())}')
```

```python
import os
os.environ['MODEL_NAME'] = 'microsoft/Phi-3-mini-4k-instruct'
os.environ['TRAINING_OUTPUT_DIR'] = 'training/checkpoints'

from training.reward_trainer import train, ModelConfig, TrainingConfig

model_cfg = ModelConfig()
train_cfg = TrainingConfig(
    num_train_epochs=5,
    gradient_accumulation_steps=16,
    max_length=1024,
    learning_rate=1e-4,
)

final_dir = train(data_path='data', model_cfg=model_cfg, train_cfg=train_cfg)
print(f'Training complete: {final_dir}')
```
